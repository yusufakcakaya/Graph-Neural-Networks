```python
#Import necessary libraries 
import torch
from torch_geometric.data import Data
import networkx as nx
from torch_geometric.utils.convert import to_networkx, from_networkx

```

Graph is a collection of nodes and the edges between the nodes.

## Workshop graph

```python
edge_list = torch.tensor([
                         [0, 0, 0, 1, 2, 2, 3, 3], # Source Nodes
                         [1, 2, 3, 0, 0, 3, 2, 0]  # Target Nodes
                        ], dtype=torch.long)
```

```python
node_features = torch.tensor([
                            [-8, 1, 5, 8, 2, -3], # Features of Node 0
                            [-1, 0, 2, -3, 0, 1], # Features of Node 1
                            [1, -1, 0, -1, 2, 1], # Features of Node 2
                            [0, 1, 4, -2, 3, 4], # Features of Node 3
                            ],dtype=torch.long)
```

```python
# Make a data object to store graph informaiton 
data = Data(x=node_features, edge_index=edge_list)
```

```python
# Draw data with networkx
G = to_networkx(data)
nx.draw_networkx(G)
```
